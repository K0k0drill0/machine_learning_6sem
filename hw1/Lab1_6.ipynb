4. Как найти самое часто встречаемое значение между двумя массивами?

````python
import numpy as np

array1 = np.array([1, 1, 3, 2, 2, 2, 2, 2])
array2 = np.array([1, 3, 1, 2])

common_elements = np.intersect1d(array1, array2)

filtered_array1 = np.array([x for x in array1 if x in common_elements])
filtered_array2 = np.array([x for x in array2 if x in common_elements])

overall_combined_array = np.concatenate((filtered_array1, filtered_array2))

values, counts = np.unique(overall_combined_array, return_counts=True)

most_frequent_value = values[np.argmax(counts)]

print(most_frequent_value)
````

8. Создайте случайную матрицу 10х2 которая представляет собой декартовые координаты и преобразуйте их в полярные

````python
import numpy as np
import random as rand

cartesian_coords = np.random.rand(10, 2) * 10

X = cartesian_coords[:, 0]
Y = cartesian_coords[:, 1]

r = np.sqrt(X**2 + Y**2)
theta = np.arctan2(Y, X)

polar_coordinates = np.column_stack((r, theta))
print(cartesian_coords)
print(polar_coordinates)
````

3. Как зависит результат конечного экзамена от результата первого экзамена

````python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(
    data=df,
    x='Previous_Scores',
    y='Exam_Score',
    line_kws={'color': 'red'}
)
plt.title('Зависимость конечного балла от начального')
plt.xlabel('Балл за первый экзамен')
plt.ylabel('Балл за конечный экзамен')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()
````

4. Постройте box plot отображающий распределение количества сна детей сдавших экзамены на оценку ниже и выше 85 баллов

````python
df['Score_Group'] = df['Exam_Score'].apply(lambda x: '>= 85' if x >= 85 else '< 85')

plt.figure(figsize=(8, 6))
sns.boxplot(
    x='Score_Group',
    y='Sleep_Hours',
    hue='Score_Group',
    data=df,
    palette='pastel',
    legend=False
)
plt.title('Распределение часов сна в зависимости от оценки за экзамен')
plt.xlabel('Группа по баллу на экзамене')
plt.ylabel('Часы сна в день')
plt.grid(True)
plt.show()
````

7. Постройте HeatMap где ось Х - процент посещаемость, Y - количество занятий с репетитором, а Z - балл за итоговый экзамен*
